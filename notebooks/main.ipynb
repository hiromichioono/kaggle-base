```python
import pandas as pd
import numpy as np
```

```python
# lags.parquetファイルのパスを指定
lags_file_path = 'jane-street-real-time-market-data-forecasting/lags.parquet/date_id=0'

# Parquetファイルをpyarrowを使って読み込む
df = pd.read_parquet(lags_file_path, engine='pyarrow')

# カラムを省略せずに表示
pd.set_option('display.max_columns', None)
# 行を省略せずに表示
pd.set_option('display.max_rows', None)

# dfの概要を表示
print("lags.parquetのサイズ:", df.shape)
print("dfの先頭5行:\n", df.head())
print("lags.parquetの列名:\n", df.columns)

# 統計量を表示(jupyter notebookのみで表示)
df.describe()
```

```python
# lags.parquetのnan確認
for col in df.columns:
    print(col)
    print(df[col].unique())
```
